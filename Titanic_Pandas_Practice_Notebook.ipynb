Functions to explore:

### df.assign()

Crea columnas en el DataFrame sin modificar el original delvolviendo una copia del mismo

```python
# Ejemplo df.assign()
df2 = df.assign(Age_2 = df["Age"] * 2)
df2.head()
```

```python
# Ejemplo 2
df2 = df.assign(Fare_per_Age = df["Fare"] / df["Age"])
df2.head()
```

### df.apply()

Aplica funciones a cada elemento, columna o fila del DataFrame

```python
# Ejemplo df.apply()
df["Age_group"] = df["Age"].apply(lambda x: "Young" if x < 18 else "Adult")
df["Age_group"]
```

```python
# Ejemplo 2
df["Total_Family"] = df.apply(lambda row: row["SibSp"] + row["Parch"], axis=1)
df["Total_Family"]
```

### df["new_col"]

Crea o modifica una columna del DataFrame original

```python
# Ejemplo df["new_col"]
df["Fare_per_Age"] = df["Fare"] / df["Age"]
df[["Age", "Fare", "Fare_per_Age"]].head()
```

```python
# Ejemplo 2
df["Ship"] = "Titanic"
df["Ship"]
```

### pd.concat()

Une DataFrames por filas o columnas

```python
# Ejemplo pd.concat()
df_combined = pd.concat([df[["Name"]], df[["Age"]]], axis=1)
df_combined
```

```python
# Ejemplo 2
df_test = pd.DataFrame({
    "PassengerId": [4, 5],
    "Survived": [1, 0],
    "Pclass": [1, 3],
    "Name": ["Futrelle, Mrs. Jacques", "Allen, Mr. William"],
    "Sex": ["female", "male"],
    "Age": [35, 35],
    "Fare": [53.1, 8.05]
})
df_train = pd.DataFrame({
    "PassengerId": [1, 2, 3],
    "Survived": [0, 1, 1],
    "Pclass": [3, 1, 3],
    "Name": ["Braund, Mr. Owen", "Cumings, Mrs. John", "Heikkinen, Miss. Laina"],
    "Sex": ["male", "female", "female"],
    "Age": [22, 38, 26],
    "Fare": [7.25, 71.2833, 7.925]
})
df_total = pd.concat([df_train, df_test])
df_total.head()
```

### pd.merge()

Combina dos DataFrames  por una columna comun (Join en SQL)

```python
#Ejemplo pd.merge
df_personal = df[["PassengerId", "Name", "Sex", "Age", "Pclass"]]
df_supervivencia = df[["PassengerId", "Survived"]]

df_final = pd.merge(df_personal, df_supervivencia, on="PassengerId", how="inner")
df_final.head()
```

```python
# Ejemplo 2
df_economico = df[["PassengerId", "Ticket", "Fare", "Cabin", "Embarked"]]
df_final = pd.merge(df_personal, df_economico, on="PassengerId", how="inner")
df_final.head()
```
